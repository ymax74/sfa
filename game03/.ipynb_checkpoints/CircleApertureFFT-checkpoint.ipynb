Дифракция на круглом отверстии через преобразование Фурье

```python
import numpy as np
from math import *

from matplotlib import cm
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
def circular_aperture(N=1000, ap_r=100):

    u = np.linspace(-1, 1, N)
    x, y = np.meshgrid(u, u)
    R = np.sqrt(x**2 + y**2)
    
    rho = R / (ap_r / (N/2))
    ph = np.where(rho <= 1.0, 1.0, 0.0)
    
    U = np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(ph)))
    I = np.abs(U)**2
    psf = I/np.max(I)  

    return ph, psf
```

```python
ap, psf = circular_aperture()

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(ap, cmap='gray')
plt.subplot(122)
plt.imshow(np.log10(psf + 1e-6), cmap='magma') 
```

```python
def circular_aperture2(N=1000, ap_r=60, incl = 0.0, defp=0.0, sph =0.0, dist=0.0 ,coma=0.0, astig=0.0):

    u = np.linspace(-1, 1, N)
    x, y = np.meshgrid(u, u)
    R = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    
    rho = R / (ap_r / (N/2))
    ph = np.where(rho <= 1.0, 1.0, 0.0)
    
#  наклон
#     ph_tilt = sin(radians(theta)) * x * (N / ap_r)
    ph_tilt = incl * x * (N / ap_r)
#  дефокусировка
    ph_def = defp*(2*rho**2-1)
#  сферическая аберрация
    ph_sph = sph*(6*rho**4-6*rho**2+1)
#  дисторсия
    ph_dist = dist*rho*np.cos(phi)
#  кома 
    ph_coma = coma * (3*rho**3-2*rho) * np.cos(phi)
#  астигматизм 
    ph_astig = astig * rho**2*np.cos(2*phi)
    
    total_ph = ph_tilt + ph_def + ph_sph + ph_dist + ph_coma + ph_astig
    U_in = ph * np.exp(1j * total_ph)
    
    U_out = np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(U_in)))
    I = np.abs(U_out)**2
    psf = I/np.max(I)

    return ph, total_ph, psf
```

```python
ph, total_ph, psf = circular_aperture2(incl=0.0, defp=0.0, sph=0.0, dist=0.0, coma=0.0, astig=0.0)
```

```python
def plot(ph, total_ph, psf):
    plt.figure(figsize=(8, 4))
    plt.subplot(121)
    plt.imshow(ph*total_ph, cmap='gray')
    plt.subplot(122)
    plt.imshow(np.log10(psf + 1e-6), cmap='magma') 
    plt.axvline(500, color='white', linestyle='--', alpha=0.3) # Оптическая ось
    plt.axhline(500, color='white', linestyle='--', alpha=0.3)
```

```python
def centroid(data: np.ndarray):
    h, w = data.shape
    x_idx = np.arange(w)
    #y_idx = np.arange(h)
    x_sum = np.sum(data, axis=0)
    #y_sum = np.sum(data, axis=1)
    
    x_c = np.sum(x_sum*x_idx) / np.sum(x_idx)
    #y_c = np.sum(y_sum*y_idx) / np.sum(y_idx)
    
    return x_c, 0
    
```

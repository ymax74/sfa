```python
# rad
def dist(ra, dec, RA, DEC):
    cos_rho = np.sin(dec)*np.sin(DEC)+np.cos(dec)*np.cos(DEC)*np.cos(ra-RA)
    if 1.0<cos_rho<1.0+1e-6:
        return 0.0 # костыль для ошибок округления где результат чуть больше 1 
    if -1.0>cos_rho>-1.0+1e-6:
        return np.pi # костыль для ошибок округления где результат чуть больше 1 
    return np.arccos(cos_rho)
```

```python
# rad
def sph2tang(ra, dec, RA, DEC):
    cosrho = np.sin(dec)*np.sin(DEC)+np.cos(dec)*np.cos(DEC)*np.cos(ra-RA)
    xi = np.cos(dec)*np.sin(ra-RA)/cosrho
    eta = (np.sin(dec)*np.cos(DEC)-np.cos(dec)*np.sin(DEC)*np.cos(ra-RA))/cosrho
    return xi,eta
```

```python
# rad
def tang2sph(xi, eta, RA, DEC):
    T = np.array([[-sin(RA), -cos(RA)*sin(DEC),   cos(RA)*cos(DEC)],
                  [ cos(RA), -sin(RA) * sin(DEC), sin(RA) * cos(DEC)],
                  [       0,  cos(DEC),           sin(DEC)]])
    x,y,z = np.dot(T, [xi,eta,1])
    ra = atan2(y,x)
    dec = atan2(z, np.sqrt(x*x+y*y))
    if(ra<0):
        ra+=2*np.pi
    return ra,dec
```

Простой тест:

```python
ra1 = RA1 * np.pi/180.0
dec1 = DEC1 * np.pi/180.0
print(ra1-0.1, dec1-0.1)

test_xi1, test_eta1 = sph2tang(ra1-0.1, dec1-0.1, ra1, dec1)
print(test_xi1, test_eta1)
test_ra1, test_dec1 = tang2sph(test_xi1, test_eta1, ra1, dec1)
print(test_ra1, test_dec1)

print()

ra2 = RA2* np.pi/180.0
dec2= DEC2* np.pi/180.0
print(ra2-0.1, dec2-0.1)

test_xi2, test_eta2 = sph2tang(ra2-0.1, dec2-0.1, ra2, dec2)
print(test_xi2, test_eta2)
test_ra2, test_dec2 = tang2sph(test_xi2, test_eta2, ra2, dec2)
print(test_ra2, test_dec2)

print()

eps = 1e-6
print(np.hypot(test_ra1 - ra1 + 0.1, test_dec1 - dec1 + 0.1))
print(np.hypot(test_ra2 - ra2 + 0.1, test_dec2 - dec2 + 0.1))
```

```python
assert np.hypot(test_ra1 - ra1 + 0.1, test_dec1 - dec1 + 0.1) <= eps
assert np.hypot(test_ra2 - ra2 + 0.1, test_dec2 - dec2 + 0.1) <= eps
```

```python
# deg
def advance_time_naive(ra, dec, pmra, pmdec, t):
    return (ra+pmra*t, dec+pmdec*t)
```

```python
# deg
def advance_time_tangential(ra, dec, pmra, pmdec, t):
    # tang2sph принимает радианы, конвертируем из градусов
    ra2, dec2 = tang2sph(pmra*t*np.pi/180.0, pmdec*t*np.pi/180.0, ra*np.pi/180.0, dec*np.pi/180.0)
    return ra2/np.pi*180.0, dec2/np.pi*180.0 # возвращаем в градусах
```

```python
years = 1000

pos_naive1 = []
pos_tang1 = []
for idx, row in df1.iterrows():
    ra = row["ra"]
    dec = row["dec"]
    pmra = row["pmra"] / 3.6e6
    pmdec = row["pmdec"] / 3.6e6
    
    posn = advance_time_naive(ra, dec, pmra, pmdec, years)
    post = advance_time_tangential(ra, dec, pmra, pmdec, years)
    pos_naive1.append(posn)
    pos_tang1.append(post)
    
pos_naive2 = []
pos_tang2 = []
for idx, row in df2.iterrows():
    ra = row["ra"]
    dec = row["dec"]
    pmra = row["pmra"] / 3.6e6
    pmdec = row["pmdec"] / 3.6e6
    
    posn = advance_time_naive(ra, dec, pmra, pmdec, years)
    post = advance_time_tangential(ra, dec, pmra, pmdec, years)
    pos_naive2.append(posn)
    pos_tang2.append(post)
```

```python
dist_deg = lambda ra1, dec1, ra2, dec2: degrees(
    dist(radians(ra1), radians(dec1), radians(ra2), radians(dec2))
)

delta1 = [dist_deg(*(pos_naive1[i]), *(pos_tang1[i])) for i in range(len(pos_naive1))]
delta2 = [dist_deg(*(pos_naive2[i]), *(pos_tang2[i])) for i in range(len(pos_naive2))]
```

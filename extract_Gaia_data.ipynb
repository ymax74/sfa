### Добыча данных из каталога Gaia
Полноценное впечатление об этом процессе можно получить собственное на <a href=https://gea.esac.esa.int/archive/ target=_blank>сайте</a> архива Gaia.

```python
import numpy as np
import pandas as pd

from math import *

import ssl
# Это модуль для добычи данных из каталогов миссии Gaia
from astroquery.gaia import Gaia
```

```python
RA,DEC = 34.0,70.0
fov = 0.05
job = Gaia.launch_job_async("SELECT * FROM gaiadr3.gaia_source \
        WHERE CONTAINS(POINT('ICRS',gaiadr3.gaia_source.ra,gaiadr3.gaia_source.dec),CIRCLE('ICRS',%f,%f,%f))=1"\
                %(RA,DEC,fov), dump_to_file=True, output_format='csv',
                            output_file='Gaia_data.csv')
```

Описание таблиц Gaia рекомендуется изучать <a href=https://gea.esac.esa.int/archive/documentation/GDR3/Gaia_archive/chap_datamodel/ target=_blank>здесь</a> 

```python
# грузим в "датафрейм" табличку, добытую выше из базы данных Gaia
df = pd.read_csv('Gaia_data.csv') 
# вывод титулов колонок
for c in df.columns:
    print(c)
df# вывод таблички на экран для проверки
```

Предыдущий пример очень простой. Но иногда надо произвести выборку по определенным условиям (например, отобрать звезды в каком-то диапазоне блеска). Часто нелогично устраивать хранение данных в файле, пусть себе лежать в каком-нибудь массиве. Следующий пример акцентрирует внимание на этом.

```python
limmag1,limmag2 = 14,16.5
ssl._create_default_https_context = ssl._create_unverified_context
job = Gaia.launch_job_async("SELECT * \
FROM gaiaedr3.gaia_source \
WHERE CONTAINS(POINT('ICRS',gaiaedr3.gaia_source.ra,gaiaedr3.gaia_source.dec),CIRCLE('ICRS',%f,%f,%f))=1\
                           AND  phot_g_mean_mag>%f AND  phot_g_mean_mag<%f;"%(RA,DEC,fov,limmag1,limmag2) \
                            , dump_to_file=False)

gaiat = job.get_results()
gaiat# таблица с даными
```
